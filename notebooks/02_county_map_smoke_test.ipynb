```python
import pandas as pd

df = pd.DataFrame({
    "fips": ["06075", "36061", "17031"],  # San Francisco CA, New York County NY, Cook County IL
    "value": [10.5, 7.2, 12.1]            # fake numbers (pretend "uninsured %")
})

# Make sure FIPS is a 5-character string
df["fips"] = df["fips"].astype(str).str.zfill(5)

df
```

```python
import json

geojson_path = "../data/raw/geo/geojson-counties-fips.json"

with open(geojson_path, "r") as f:
    counties = json.load(f)

type(counties), counties.keys()
```

```python
import plotly.express as px

fig = px.choropleth(
    df,
    geojson=counties,
    locations="fips",
    color="value",
    scope="usa",
    color_continuous_scale="Blues",
    labels={"value": "Example value"},
)

fig.update_layout(
    title="County map smoke test (3 counties)",
    margin={"r":0,"t":50,"l":0,"b":0},
)

fig.show()
```
